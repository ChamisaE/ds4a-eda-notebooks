# Analyzing Rows 11-20

```python
#creating new df with just rows 11-20
dfv = df[["geoid",
          "travel_time_total",
          "travel_time_10_less",
          "travel_time_10_14",
          "travel_time_15_19",
          "travel_time_20_24",
          "travel_time_25_29",
          "travel_time_30_34",
          "travel_time_35_44",
          "travel_time_45_59",
          "travel_time_60_plus"]]
```

```python
dfv.head()
```

```python
dfv.fillna(0)
```

```python
#creating a new column called travel_less_than_30, and combining all of the
#columns with travel below 30 min together to populate new column.
dfv["travel_less_than_30"] = df["travel_time_10_less"] + df["travel_time_10_14"] + df["travel_time_15_19"] + df["travel_time_20_24"] + df["travel_time_25_29"]

```

```python
#creating a new column called travel_30_to_59, and combining all of the
#columns with travel within range together to populate new column.
dfv["travel_30_to_59"] = df["travel_time_30_34"] + df["travel_time_35_44"] + df["travel_time_45_59"]
```

```python
#initial correlation matrix; with new columns
corr = dfv.corr()
corr.style.background_gradient(cmap="coolwarm").set_precision(2)
```

```python
#reduce df to travel time columns

df2 = dfv[['geoid','travel_time_total','travel_less_than_30','travel_30_to_59','travel_time_60_plus']]
```

```python
df2
```

```python
#correlation matrix; with only reduced columns
corr = df2.corr()
corr.style.background_gradient(cmap="coolwarm").set_precision(2)
```

Find % of people traveling for each range

```python
df2['%_travel_less_than_30'] = df2['travel_less_than_30'] / df2['travel_time_total']
```

```python
df2
```

```python
df2['%_travel_30_to_59'] = df2['travel_30_to_59'] / df2['travel_time_total']
```

```python
df2
```

```python
df2['%_travel_over_60'] = df2['travel_time_60_plus'] / df['travel_time_total']
```

```python
df2
```
